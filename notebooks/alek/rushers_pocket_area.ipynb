Import List and Dict so I can annotate types for function parameters.

Import ConvexHull class to calculate pocket area.

```python
from typing import (List, Dict)
from scipy.spatial import ConvexHull
```

Function to calculate pocket area using the locations of the blockers as the vertices.

Parameters:
> players: A list of player dictionaries. Each player contains a role in the play (passer, blocker, rusher), x coordinate, and y coordinate.

Returns:
> The pocket area as a float.



```python
def rushers_pocket_area(players: List[Dict]) -> float:

  points = [(p['x'], p['y']) for p in players if p['role'] == 'rusher']
  hull = ConvexHull(points)
  return hull.volume
```

Example list of player dictionaries.

```python
frame = [
  {'role': 'passer', 'x': 0, 'y': 0},
  {'role': 'rusher', 'x': 1, 'y': 1},
  {'role': 'rusher', 'x': 1, 'y': -1},
  {'role': 'rusher', 'x': -1, 'y': -1},
  {'role': 'rusher', 'x': -1, 'y': 1},
  {'role': 'rusher', 'x': 0, 'y': 2},
  {'role': 'rusher', 'x': 0.5, 'y': 0.5},
  {'role': 'blocker', 'x': 2, 'y': 1},
  {'role': 'blocker', 'x': 2, 'y': -1},
  {'role': 'blocker', 'x': -2, 'y': -1},
  {'role': 'blocker', 'x': -2, 'y': 1},
  {'role': 'blocker', 'x': 0, 'y': 3}
]
```

Example function call.

```python
area = rushers_pocket_area(frame)
print(area)
```
